```python
# üåßÔ∏è Smart Rain Detection

from transformers import AutoImageProcessor, AutoModelForImageClassification
from PIL import Image
import torch

# (This model can classify images as sunny, rainy, cloudy, snowy, etc.)
model_name = "methane6923/weather_classification_ViT"

print(f"Loading model: {model_name} ...")
processor = AutoImageProcessor.from_pretrained(model_name)
model = AutoModelForImageClassification.from_pretrained(model_name)
print("Model loaded successfully ‚úÖ\n")

# -------- 2Ô∏è‚É£  Load an image ------------
# Replace this path with your own image (rainy or clear scene)
image_path = "/content/no_rain_sample.jpg"

try:
    image = Image.open(image_path)
except FileNotFoundError:
    raise SystemExit(f"‚ùå Could not find image at {image_path}. Place an image there first.")

# -------- 3Ô∏è‚É£  Preprocess and run inference ------------
inputs = processor(images=image, return_tensors="pt")
outputs = model(**inputs)
predictions = torch.nn.functional.softmax(outputs.logits, dim=-1)
predicted_class_idx = predictions.argmax(-1).item()
label = model.config.id2label[predicted_class_idx]
confidence = predictions[0][predicted_class_idx].item() * 100

print(f"üñºÔ∏è  Image: {image_path}")
print(f"üå¶Ô∏è  Predicted weather: {label} ({confidence:.2f}% confidence)\n")

# -------- 4Ô∏è‚É£  Simulate motor control logic ------------
if "rain" in label.lower():
    print("üíß Rain detected ‚Üí Turning OFF irrigation motor.")
else:
    print("‚òÄÔ∏è  No rain detected ‚Üí Motor remains ON.")

# -------- 5Ô∏è‚É£  Done! ------------
print("\n‚úÖ Process completed.")
```
